1-importar imagem

```python
from google.colab import files
from PIL import Image
from IPython.display import display


# Upload da imagem (escolha o arquivo do seu PC)
uploaded = files.upload()

# Abre a imagem (precisa ter exatamente o mesmo nome do arquivo enviado)
imagem = Image.open("ursinho.png")

# Converte a imagem para RGB (remove canal alfa, significa a trasparência, se tiver)
imagem = imagem.convert("RGB")




```

2-Converter para escala de cinza

```python

# Converter para escala de cinza
largura, altura = imagem.size
imagem_cinza = Image.new("L", (largura, altura))


# Converter pixel a pixel
for y in range(altura):
    for x in range(largura):
        r, g, b = imagem.getpixel((x, y))
        valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b)
        imagem_cinza.putpixel((x, y), valor_cinza)

print("Conversão para escala de cinza concluída!")


```

3-Binarizar a imagem

```python
def binarizar_imagem(imagem_cinza, limiar=127):
    """
    Converte uma imagem em escala de cinza para binária (preto e branco).
    """
    largura, altura = imagem_cinza.size
    imagem_binaria = Image.new("1", (largura, altura))  # Binária
    for y in range(altura):
        for x in range(largura):
            valor_cinza = imagem_cinza.getpixel((x, y))
            valor_binario = 1 if valor_cinza > limiar else 0
            imagem_binaria.putpixel((x, y), valor_binario)
    return imagem_binaria

# Binarizar a imagem
imagem_binaria = binarizar_imagem(imagem_cinza, limiar=127)
print("Binarização concluída!")
```

4 – Exibir todas as imagens

```python
from IPython.display import display

print("Imagem original:")
display(imagem)

print("Imagem em escala de cinza:")
display(imagem_cinza)

print("Imagem binarizada:")
display(imagem_binaria)
```
